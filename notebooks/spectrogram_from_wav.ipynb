```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import os
from matplotlib.image import imread
```

```python
file_path = '../data/samples/snare/snare_5.wav'
y, sr = librosa.load(file_path, sr=None)
S = librosa.stft(y)
S_dB = librosa.amplitude_to_db(np.abs(S), ref=np.max)

# Create output directory if it doesn't exist
os.makedirs("../out", exist_ok=True)

output_name = os.path.splitext(os.path.basename(file_path))[0] + '.jpg'
output_path = os.path.join('../out', output_name)

# Plot without labels or axes
fig = plt.figure(figsize=(10, 4))
ax = fig.add_subplot(111)
librosa.display.specshow(S_dB, sr=sr, cmap='magma', ax=ax)
ax.axis('off')  # Removes axes and ticks
plt.savefig(output_path, dpi=300, bbox_inches='tight', pad_inches=0)
plt.close()

print(f"Saved clean spectrogram to {output_path}")
```

```python
img = imread(output_path)
plt.figure(figsize=(10, 4))
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
# Load the audio file
file_path = '../data/samples/snare/snare_5.wav'
y, sr = librosa.load(file_path, sr=None)

# Create the Short-Time Fourier Transform (STFT)
S = librosa.stft(y)
S_dB = librosa.amplitude_to_db(np.abs(S), ref=np.max)

# Create output directory if it doesn't exist
os.makedirs("../out", exist_ok=True)

# Create output filename
output_name = os.path.splitext(os.path.basename(file_path))[0] + '.jpg'
output_path = os.path.join('../out', output_name)

# Plot and save the spectrogram
plt.figure(figsize=(10, 4))
librosa.display.specshow(S_dB, sr=sr, x_axis='time', y_axis='hz', cmap='magma')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
plt.tight_layout()
plt.savefig(output_path, dpi=300, bbox_inches='tight', pad_inches=0.1)
plt.close()

print(f"Saved spectrogram to {output_path}")
```

```python
img = imread(output_path)
plt.figure(figsize=(10, 4))
plt.imshow(img)
plt.axis('off')
plt.show()
```
